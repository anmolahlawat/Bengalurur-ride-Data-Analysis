```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['avg_vtat'] = df['avg_vtat'].fillna(df['avg_vtat'].median())
```

```python
df['avg_ctat'] = df['avg_ctat'].fillna(df['avg_ctat'].median())
```

```python
df.isnull().sum()
```

```python
df["cancelled_rides_by_driver"]
```

```python
df['reason_for_cancelling_by_customer'] = (
    df['reason_for_cancelling_by_customer']
    .fillna('Not Cancelled')
)

```

```python
df['reason_for_cancelling_by_customer']
```

```python
df.isnull().sum()
```

```python
df["cancelled_rides_by_driver"] = (
    df['cancelled_rides_by_driver']
    .fillna('Not Cancelled')
)
```

```python
df.isnull().sum()
```

```python
df["booking_value"] = df["booking_value"].fillna(df["booking_value"].mean())
df["ride_distance"] = df["ride_distance"].fillna(df["ride_distance"].mean())
df["driver_ratings"] = df["driver_ratings"].fillna(df["driver_ratings"].mean())
df["customer_rating"] = df["customer_rating"].fillna(df["customer_rating"].mean())
```

```python
df.isnull().sum()
```

```python
df['incomplete_rides_reason'] = (
    df['incomplete_rides_reason']
    .fillna('Complete')
)
```

```python

from sqlalchemy import create_engine
from urllib.parse import quote_plus


# MySQL connection details
username = "root"
password = quote_plus("Root@1234")  # Encode special characters
host = "localhost"
port = 3306
database = "Bengluru_ride_Booking"

# Create connection
engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
)

# Test connection
with engine.connect() as conn:
    print("Connected to MySQL successfully!")

# Write DataFrame to MySQL
table_name = "ride_data"

df.to_sql(
    table_name,
    con=engine,
    if_exists="replace",
    index=False
)

print(f"Table '{table_name}' created successfully!")

# Read sample rows back from MySQL
sample_df = pd.read_sql(
    f"SELECT * FROM {table_name} LIMIT 5",
    con=engine
)

print(sample_df)
```
